## amator solution

```python
# assign symbols x, y, z

x, y, z = symbols('x, y, z')
```

```python
# assign MSE(x, y, z)

MSE = 1/6 * ((5*x + 7*y - 5*z + 47)**2 + \
             (-2*y + 2*z - 10)**2 + \
             (-4*x - 8*y - 7*z - 63)**2 + \
             (x + y + 2*z + 1)**2 + \
             (2*x - y + 2*z + 4)**2 + \
             (4*x + y + 4*z + 2)**2)
MSE
```

```python
# calculate MSE in points (0, 0)
MSE.subs({x: 0, y: 0, z: 0})
```

```python
# calculate private derivatives for MSE

MSEx, MSEy, MSEz = diff(MSE, x), diff(MSE, y), diff(MSE, z)
MSEx
```

```python
# assign gradient step

step = 0.01
```

```python
# assign first point

point = np.array([0, 0, 0])
point
```

```python
# assign gradient 

grad = np.array([MSEx.subs({x: point[0], y: point[1], z: point[2]}), 
                 MSEy.subs({x: point[0], y: point[1], z: point[2]}), 
                 MSEz.subs({x: point[0], y: point[1], z: point[2]})]
               )
grad
```

```python
# calculate next point

next_point = point - step * grad
next_point
```

```python
# check MSE 

MSE.subs({x: next_point[0], y: next_point[1], z: next_point[2]})
```

```python
# assign next point

point = next_point
point
```

## other solution

```python
x, y, z = symbols('x, y, z')                                                           # assign symbols x, y, z


MSE = 1/6 * ((5*x + 7*y - 5*z + 47)**2 + \
             (-2*y + 2*z - 10)**2 + \
             (-4*x - 8*y - 7*z - 63)**2 + \
             (x + y + 2*z + 1)**2 + \
             (2*x - y + 2*z + 4)**2 + \
             (4*x + y + 4*z + 2)**2)                                                   # assign MSE function

point, step = np.array([0, 0, 0]), 0.01                                                # assign start point and gradient step
MSEx, MSEy, MSEz = diff(MSE, x), diff(MSE, y), diff(MSE, z)                            # calculate private MSE derivatives

# create while cycle for MSE < 55 searching

iter_count = 0

while True:
    
    grad = np.array([MSEx.subs({x: point[0], y: point[1], z: point[2]}),               # calculate gradient
                 MSEy.subs({x: point[0], y: point[1], z: point[2]}), 
                 MSEz.subs({x: point[0], y: point[1], z: point[2]})]
               )
    
    next_point = point - step * grad                                      
    
    mse_result = MSE.subs({x: next_point[0], y: next_point[1], z: next_point[2]})      # calculate MSE values in points
    
    point = next_point                                                                 # assign new start point
    
    iter_count += 1                                                                    # iteration calculation
    
    if mse_result < 55:                                                                # break cycle and print values
        print(f"MSE - {mse_result:.2f}\n"
              f"iteration count - {iter_count}\n"
              f"x, y, z - {next_point}")
        break
```
