```python
from sympy import *
```

```python
x, y, z = symbols('x, y, z')
```

```python
X = Matrix([[2, 5, 11], [7, 3, -23]])
X
```

```python
linsolve(X, [x, y])
```

```python
X1 = Matrix([[2, 3, 12], [3, -1, 7]])
linsolve(X1, [x, y])
```

```python
X2 = Matrix([[1, 1, 2, -1], [2, -1, 2, -4], [4, 1, 4, -2]])
X2
```

```python
linsolve(X2, [x, y, z])
```

```python
A = Matrix([[2, 5], [7, -3]])
b = Matrix([2, 3])
A * b
```

```python
A1 = Matrix([[-3, 1], [3, 5]])
b1 = Matrix([1, 3])
A1 * b1
```

```python
A2 = Matrix([[2, 3], [3, -1]])
b2 = Matrix([4, 2])
A2 * b2
```

```python
A3 = Matrix([[1, 1, 2], [2, -1, 2], [4, 1, 4]])
b3 = Matrix([2, 3, 5])
A3 * b3
```
