# Задание 2. Псевдообратная матрица
Решите задачу аппроксимации из практической работы 9 модуля (задание 3) с помощью псевдообратной матрицы (а не градиентного спуска), используя для вычислений SymPy. Насколько точнее получилось решение?

```python
# assign matrices and symbols

a1, a2 = symbols('a1, a2')

# assign MSE eqation 

MSE = 1/4 * ((1*a1 + 2*a2 - 5)**2 + \
             (5*a1 + 3*a2 - 6)**2 + \
             (2*a1 + 4*a2 - 10)**2 + \
             (3*a1 + 7*a2 - 8)**2)

# assign matrices 

X = Matrix([[1, 2], [5, 3], [2, 4], [3, 7]])
b = Matrix([5, 6, 10, 8])

# assign pseudo matrix equation

m = (X.T * X)**(-1) * X.T

# calculate points 

point = m * b

point
```

```python
# calculate MSE by points

MSE.subs({a1: point[0], a2: point[1]})
```
