```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality

init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
# task 1

# assign variables and calculate all coefficients for polynomials in fourth degree

a4, a3, a2, a1, a0 = symbols('a4, a3, a2, a1, a0')                                   # assign variables for coefficients

eq1_lp = a4*(-2)**4 + a3*(-2)**3 + a2*(-2)**2 + a1*(-2) + a0 - 2                     # solve the equations
eq2_lp = a4*(-1)**4 + a3*(-1)**3 + a2*(-1)**2 + a1*(-1) + a0 - 1 
eq3_lp = a4*(0)**4 + a3*(0)**3 + a2*(0)**2 + a1*(0) + a0 - 0
eq4_lp = a4*(1)**4 + a3*(1)**3 + a2*(1)**2 + a1*(1) + a0 - 1
eq5_lp = a4*(2)**4 + a3*(2)**3 + a2*(2)**2 + a1*(2) + a0 - 2

nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp], [a4, a3, a2, a1, a0])          # use nonlinsolve function
```

```python
# assign x
x = Symbol('x')
```

```python
# write function equation
f = -1/6*x**4 + 0*x**3 + 7/6*x**2 + 0*x + 0
```

```python
# check for parity
f.subs(x, 1), f.subs(x, -1)
```

```python
# find zeros
solve(f)
```

```python
# find constancy intervals
solve_univariate_inequality(f>0, x), solve_univariate_inequality(f<0, x)
```

```python
# create plot
plot(f)
```

```python
# task 2 (−3,33),(−2,31),(−1,18),(1,−18),(2,−31),(3,−33)

# # assign variables and calculate all coefficients for polynomials in fourth degree

a5, a4, a3, a2, a1, a0 = symbols('a5, a4, a3, a2, a1, a0')                              # assign variables for coefficients

eq1_lp = a5*(-3)**5 + a4*(-3)**4 + a3*(-3)**3 + a2*(-3)**2 + a1*(-3) + a0 - 33          # solve the equation
eq2_lp = a5*(-2)**5 + a4*(-2)**4 + a3*(-2)**3 + a2*(-2)**2 + a1*(-2) + a0 - 31
eq3_lp = a5*(-1)**5 + a4*(-1)**4 + a3*(-1)**3 + a2*(-1)**2 + a1*(-1) + a0 - 18
eq4_lp = a5*(1)**5 + a4*(1)**4 + a3*(1)**3 + a2*(1)**2 + a1*(1) + a0 + 18
eq5_lp = a5*(2)**5 + a4*(2)**4 + a3*(2)**3 + a2*(2)**2 + a1*(2) + a0 + 31
eq6_lp = a5*(3)**5 + a4*(3)**4 + a3*(3)**3 + a2*(3)**2 + a1*(3) + a0 + 33

nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp, eq6_lp], [a5, a4, a3, a2, a1, a0]) # use nonlinsolve function
```

```python
# assign x
x = Symbol('x')
```

```python
# write function equation
f = 1/120*x**5 + 0*x**4 + 19/24*x**3 + 0*x**2 + (-94/5*x) + 0
f
```

```python
# check for parity
f.subs(x, 1), f.subs(x, -1)
```

```python
# find zeros
solve(f)
```

```python
# find constancy intervals
print(f"{solve_univariate_inequality(f>0, x)}\n"
      f"{solve_univariate_inequality(f<0, x)}")
```
