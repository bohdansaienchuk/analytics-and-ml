```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

## 1. 

```python
f = sin(sqrt(x ** 2 + y ** 2))
f
```

```python
plot3d(f)
```

## 2. 

```python
f = exp(-0.9 * (x**2 + 0.5 * (x - y)**2))
f
```

```python
plot3d(f, (x, -3, 3), (y, -3, 3))
```

## 3. 

```python
f = x * cos(y)
f
```

```python
plot3d(f)
```
