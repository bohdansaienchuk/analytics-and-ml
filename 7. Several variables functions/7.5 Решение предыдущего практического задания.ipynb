```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

## 1. 

```python
f = 2 * x + 5 * y + 1
f
```

```python
plot3d(f);
```

```python
solve(f)
```

```python
f = - 5 * y / 2 - 1 / 2
f
```

```python
plot(f);
```

## 2. 

```python
f = x**2 + y**2 - 1
f
```

```python
plot3d(f);
```

```python
solve(f)
```

```python
f = - sqrt(1 - y**2)
f
```

```python
plot(f, (y, -1, 1));
```

```python
f = sqrt(1 - y**2)
f
```

```python
plot(f, (y, -1, 1));
```
