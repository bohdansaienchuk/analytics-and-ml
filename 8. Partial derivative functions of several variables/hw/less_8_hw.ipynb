### $f(x,y) = x(y-5)^2$

```python
# import all necessary libraries

from sympy import *
from sympy.solvers.inequalities import solve_univariate_inequality
from sympy.plotting import plot
from sympy.plotting import plot3d

import math

init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
# assign symbols x and y

x, y = symbols('x, y')
```

```python
# create function

f = x*(y-5)**2
f
```

```python
# find private derivative for x

fx = diff(f, x)
fx
```

```python
# draw graph for fx

plot(fx)
```

```python
# find zeros for fx

solve(fx)
```

```python
# find private derivative for y

fy = diff(f, y)
fy
```

```python
# draw graph for fy

plot3d(fy)
```

```python
# find zeros for fy

solve(y)
```

## Задание 2. Минимизация среднеквадратической ошибки
Найдите наилучшую, с точки зрения среднеквадратичной ошибки, функцию, приближающую зависимость времени разгона автомобиля до 100 км/ч от его массы и мощности двигателя. 

points values (1945, 560, 4.3), (1495, 340, 4.9), (1570, 343, 5.2) <br>
where x1 = 1945, x2 = 560 and y = 4.3 and the same for other points

```python
# choose function for data describing - for example it will be linear function

# f(x2, x1) = a2*x2 + a1*x1 + a0
```

```python
# create necessary symbols 

x1, x2 = symbols('x1, x2')
a0, a1, a2 = symbols('a0, a1, a2')
```

```python
# write MSE function

MSE = 1/3 * (((a2*1945 + a1*560 + a0) - 4.3)**2 +
             ((a2*1495 + a1*340 + a0) - 4.9)**2 +
             ((a2*1570 + a1*343 + a0) - 5.2)**2
            )
MSE
```

```python
# find private derivative for MSEa2, MSEa1, MSEa0

MSEa2 = diff(MSE, a2)
MSEa1 = diff(MSE, a1)
MSEa0 = diff(MSE, a0)
```

```python
# use nonlinsolve function for coefficiets calculation

nonlinsolve([MSEa2, MSEa1, MSEa0], [a2, a1, a0])
```

```python
# set coefficients for MSE check in 

MSE.subs({a2: 0.00447524752475138, a1: -0.0118811881188098, a0: 2.24910891089204})
```

```python
# set opposite value for MSE check in

MSE.subs({a2: -0.00447524752475138, a1: -0.0118811881188098, a0: 2.24910891089204})
```
